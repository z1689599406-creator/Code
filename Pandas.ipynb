```python
# 1. 创建一个包含各分组平均年龄的 Series，其索引与原始 df 完全一致
#    transform('mean') 会为 df 中的每一行找到它所属组的平均年龄，并返回
grouped_age_series = df.groupby(['Pclass', 'Sex'])['Age'].transform('mean')
grouped_age_series
```

```python
# 1. 创建一个包含各分组平均年龄的 Series，其索引与原始 df 完全一致
#    transform('mean') 会为 df 中的每一行找到它所属组的平均年龄，并返回
grouped_age_series = df.groupby(['Pclass', 'Sex'])['Age'].transform('mean')

# 2. 使用这个新创建的 Series 来填充原始 'Age' 列中的 NaN
#    .fillna() 会用 grouped_age_series 中对应索引的值来填充
#    为了不破坏原始数据，我们创建一个新列 Age_imputed
df['Age_imputed'] = df['Age'].fillna(grouped_age_series)

# 3. 验证结果
print("填充后 'Age_imputed' 列的缺失值数量:", df['Age_imputed'].isnull().sum())

# 让我们看看具体的填充效果
# .loc[] 里面传入一个条件，筛选出原始Age是NaN的行
# 然后我们只看 'Pclass', 'Sex', 'Age', 'Age_imputed' 这几列
print("\n查看部分填充效果:")
print(df.loc[df['Age'].isnull()][['Pclass', 'Sex', 'Age', 'Age_imputed']].head())
```

```python
# 假设 df 是你已经加载并清洗过的泰坦尼克数据集 DataFrame

# 第1步：定义映射规则
# 我们约定 female 为 1, male 为 0。这个约定是自定义的，只要保持一致即可。
sex_mapping = {'female': 1, 'male': 0}

# 第2步：应用映射，并创建新的一列来存放结果
# 养成创建新列的好习惯，可以避免直接修改原始数据
df['Sex_encoded'] = df['Sex'].map(sex_mapping)

# 第3步：验证结果
# 打印原始列和新创建的列的前5行进行对比
print("转换效果对比：")
print(df[['Sex', 'Sex_encoded']].head())

# 你也可以检查新列的数据类型，确认它现在是数值类型
print("\n新列的数据类型：")
df['Sex_encoded'].info()
```
